## Create function to tokenize columns and remove stop words

```python
def tokenize(col):
    tokens = [word_tokenize(content.lower()) for content in col]
    stop = set(stopwords.words('english'))
    docs = [[word for word in words if word not in stop] for words in tokens] 
    snowball_docs = [[snowball.stem(word) for word in words] for words in docs]
    return snowball_docs
```
